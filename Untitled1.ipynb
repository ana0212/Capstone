```python
from sklearn.metrics import classification_report
limiar = 5
grupos_raciais = df['race'].unique()

# Calculando as métricas para cada grupo racial
for grupo in grupos_raciais:
    grupo_df = df[df['race'] == grupo]
    
    # Convertendo o decile_score em previsões binárias para este grupo racial
    y_pred_binary = (grupo_df['decile_score'] >= limiar).astype(int)
    
    # Calculando as métricas
    y_true = grupo_df['is_recid']
    report = classification_report(y_true, y_pred_binary, output_dict=True)
    
    # Imprimindo as métricas
    print(f"Relatório de Classificação para o grupo racial '{grupo}' (limiar = {limiar}):")
    print(classification_report(y_true, y_pred_binary))
```

```python
def impact_disparity(y_true, y_pred, grupo_protegido, grupo_nao_protegido):
    # Calculando a taxa de erro para cada grupo
    taxa_erro_protegido = sum((y_true == 1) & (y_pred == 0) & (grupo_protegido == 1)) / sum(y_true == 1)
    taxa_erro_nao_protegido = sum((y_true == 1) & (y_pred == 0) & (grupo_nao_protegido == 1)) / sum(y_true == 1)
    
    # Calculando a disparidade de impacto
    impact_disparity = taxa_erro_protegido - taxa_erro_nao_protegido
    
    return impact_disparity

def statistical_parity_disparity(y_true, y_pred, grupo_protegido, grupo_nao_protegido):
    # Calculando as taxas de positividade para cada grupo
    taxa_positividade_protegido = sum(y_pred == 1) / len(y_pred)
    taxa_positividade_nao_protegido = sum(y_pred == 1) / len(y_pred)
    
    # Calculando a disparidade de estatística
    statistical_parity_disparity = taxa_positividade_protegido - taxa_positividade_nao_protegido
    
    return statistical_parity_disparity

def equal_opportunity_disparity(y_true, y_pred, grupo_protegido, grupo_nao_protegido):
    # Calculando as taxas de verdadeiros positivos para cada grupo
    taxa_verdadeiros_positivos_protegido = sum((y_true == 1) & (y_pred == 1) & (grupo_protegido == 1)) / sum(y_true == 1)
    taxa_verdadeiros_positivos_nao_protegido = sum((y_true == 1) & (y_pred == 1) & (grupo_nao_protegido == 1)) / sum(y_true == 1)
    
    # Calculando a disparidade de oportunidade igual
    equal_opportunity_disparity = taxa_verdadeiros_positivos_protegido - taxa_verdadeiros_positivos_nao_protegido
    
    return equal_opportunity_disparity

def equalized_odds_disparity(y_true, y_pred, grupo_protegido, grupo_nao_protegido):
    # Calculando as taxas de verdadeiros positivos e verdadeiros negativos para cada grupo
    taxa_verdadeiros_positivos_protegido = sum((y_true == 1) & (y_pred == 1) & (grupo_protegido == 1)) / sum(y_true == 1)
    taxa_verdadeiros_positivos_nao_protegido = sum((y_true == 1) & (y_pred == 1) & (grupo_nao_protegido == 1)) / sum(y_true == 1)
    taxa_verdadeiros_negativos_protegido = sum((y_true == 0) & (y_pred == 0) & (grupo_protegido == 1)) / sum(y_true == 0)
    taxa_verdadeiros_negativos_nao_protegido = sum((y_true == 0) & (y_pred == 0) & (grupo_nao_protegido == 1)) / sum(y_true == 0)
    
    # Calculando a disparidade de odds equalizados
    equalized_odds_disparity = abs(taxa_verdadeiros_positivos_protegido - taxa_verdadeiros_positivos_nao_protegido) + abs(taxa_verdadeiros_negativos_protegido - taxa_verdadeiros_negativos_nao_protegido)
    
    return equalized_odds_disparity
```

```python
from sklearn.metrics import precision_score, recall_score, roc_curve, roc_auc_score

# Calcular métricas de desempenho por decile
metricas_por_decile = {}
for decile in df['decile_score'].unique():
    decile_df = df[df['decile_score'] == decile]
    y_true = decile_df['is_recid']
    y_pred = (decile_df['decile_score'] >= limiar).astype(int)
    metricas_por_decile[decile] = {
        'precisao': precision_score(y_true, y_pred),
        'recall': recall_score(y_true, y_pred),
        'auc_roc': roc_auc_score(y_true, y_pred)
    }

# Plot das métricas de desempenho por decile
metricas_df = pd.DataFrame(metricas_por_decile).T
plt.figure(figsize=(10, 6))
sns.lineplot(data=metricas_df, markers=True)
plt.title('Desempenho do Modelo por Decile')
plt.xlabel('Decile Score')
plt.ylabel('Valor da Métrica')
plt.legend(metricas_df.columns)
plt.show()
```

```python
# Calcular a curva ROC por decile
fpr, tpr, thresholds = roc_curve(df['is_recid'], df['decile_score'])

# Plot da curva ROC por decile
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, marker='.')
plt.title('Curva ROC por Decile')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.grid(True)
plt.show()
```

Com base nas métricas fornecidas e nas curvas ROC (Receiver Operating Characteristic), podemos tirar algumas conclusões sobre o modelo:

Precisão e Recall: Para decile_score de 1, 2, 3 e 4, a precisão e o recall são ambos 0. Isso indica que o modelo não conseguiu prever corretamente os casos positivos e negativos para esses deciles.

Área sob a curva ROC (AUC-ROC): Para todos os decile_scores, a AUC-ROC é de 0.5. Isso significa que o modelo não é capaz de distinguir entre classes positivas e negativas melhor do que um modelo aleatório.

Precisão, Recall e AUC-ROC para decile_score de 10: Para decile_score de 10, a precisão é alta (cerca de 78%), o recall é 100% e a AUC-ROC é 0.5. Isso sugere que o modelo está tendendo a classificar a maioria dos casos como positivos, mas ainda não é capaz de distinguir entre as classes de forma significativa.

```python
def agrupar_tipo_crime(descricao):
    if pd.isna(descricao):
        return 'NA' 
    descricao = descricao.lower()
    if 'battery' in descricao or 'assault' in descricao or 'violence' in descricao or 'murder' in descricao or 'batt' in descricao:
        return 'violence'
    elif 'theft' in descricao or 'burglary' in descricao or 'robbery' in descricao:
        return 'robbery'
    elif 'drug' in descricao or 'possession' in descricao or 'trafficking' in descricao or 'poss' in descricao or 'cocaine' in descricao or 'heroin' in descricao or 'deliver' in descricao or 'traffick' in descricao:
        return 'drugs'
    elif 'driving' in descricao or 'traffic' in descricao or 'license' in descricao or 'driv' in descricao or 'vehicle' in descricao or 'conduct' in descricao:
        return 'traffic'
    else:
        return 'other'
    
df_drop_cat = df_drop.copy()   
df_drop_cat['c_charge_desc'] = df_drop_cat['c_charge_desc'].apply(agrupar_tipo_crime)
df_drop_cat['r_charge_desc'] = df_drop_cat['r_charge_desc'].replace('NA', np.nan)
df_drop_cat['r_charge_desc'] = df_drop_cat['r_charge_desc'].apply(agrupar_tipo_crime)
df_drop_cat['vr_charge_desc'] = df_drop_cat['vr_charge_desc'].replace('NA', np.nan)
df_drop_cat['vr_charge_desc'] = df_drop_cat['vr_charge_desc'].apply(agrupar_tipo_crime)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from fairlearn.reductions import ExponentiatedGradient, DemographicParity
from fairlearn.metrics import MetricFrame, selection_rate, false_positive_rate, true_positive_rate
from sklearn.calibration import CalibratedClassifierCV
import numpy as np

# Supondo que df_processed já esteja definido
categorical_features = ['sex', 'c_charge_degree', 'r_charge_degree', 'vr_charge_degree', 'c_charge_desc', 'r_charge_desc', 'vr_charge_desc']

# Definir as transformações para colunas categóricas
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combinar as transformações
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features)]
)

df_processed = df_drop_cat_dt.copy()
columns_drop = ['is_recid', 'is_violent_recid']
df_processed = df_processed.drop(columns_drop, axis=1)

# Dividir os dados em X e y
X = df_processed.drop(['two_year_recid'], axis=1)
y = df_processed['two_year_recid']

# Aplicar pré-processamento
X_transformed = preprocessor.fit_transform(X.drop(columns=['race'])).toarray()  # Convertendo para array denso

# Dividir os dados em conjunto de treinamento e teste com amostragem estratificada
X_train, X_test, y_train, y_test, train_index, test_index = train_test_split(
    X_transformed, y, df_processed.index, test_size=0.3, random_state=42, stratify=df_processed['race'])

# Encode sensitive features
sensitive_features_train = df_processed.loc[train_index]['race'].reset_index(drop=True)
label_encoder = LabelEncoder()
sensitive_features_encoded = label_encoder.fit_transform(sensitive_features_train)

# Balancear as classes minoritárias usando SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
sensitive_features_resampled = smote.fit_resample(pd.DataFrame(sensitive_features_encoded), y_train)[0].values.ravel()

# Ajustar os pesos das classes usando uma busca em grade
best_diff = float('inf')
best_weights = None
best_model = None
best_metric_frame = None

class_weights_list = [{0: 1, 1: weight} for weight in np.arange(0.1, 6.1, 0.5)]

for class_weights in class_weights_list:
    clf = GradientBoostingClassifier(random_state=42)
    clf.fit(X_train_resampled, y_train_resampled)

    # Calibrar o classificador usando validação cruzada
    calibrated_clf = CalibratedClassifierCV(base_estimator=clf, method='sigmoid', cv=5)
    calibrated_clf.fit(X_train_resampled, y_train_resampled)

    # Usando ExponentiatedGradient para mitigar o viés
    expgrad = ExponentiatedGradient(
        estimator=calibrated_clf,
        constraints=DemographicParity(),
        eps=0.01,
        max_iter=100  # Reduzir o número de iterações para melhorar a eficiência
    )
    expgrad.fit(X_train_resampled, y_train_resampled, sensitive_features=sensitive_features_resampled)
    y_pred_expgrad = expgrad.predict(X_test)

    # Avaliar o desempenho para ExponentiatedGradient
    sensitive_features_test = df_processed.loc[test_index]['race']
    metric_frame_expgrad = MetricFrame(
        metrics={'accuracy': accuracy_score, 'precision': precision_score, 'selection_rate': selection_rate,
                 'false_positive_rate': false_positive_rate, 'true_positive_rate': true_positive_rate},
        y_true=y_test,
        y_pred=y_pred_expgrad,
        sensitive_features=sensitive_features_test
    )

    # Calcular a diferença máxima na precisão entre os grupos étnicos
    precision_diff = metric_frame_expgrad.by_group['precision'].max() - metric_frame_expgrad.by_group['precision'].min()

    if precision_diff < best_diff:
        best_diff = precision_diff
        best_weights = class_weights
        best_model = expgrad
        best_metric_frame = metric_frame_expgrad

print(f"Best Class Weights: {best_weights}, Precision Difference: {best_diff:.2f}")
print("Desempenho com ExponentiatedGradient usando as melhores ponderações de classe:")
print(best_metric_frame.by_group)

# Avaliar o desempenho para cada grupo étnico com o melhor modelo
ethnic_groups = df_processed.loc[test_index]['race'].unique()
for group in ethnic_groups:
    group_indices = (df_processed.loc[test_index]['race'] == group)
    X_test_group = X_test[group_indices]
    y_test_group = y_test[group_indices]

    # Fazer previsões para o grupo
    y_pred_group = best_model.predict(X_test_group)

    if len(set(y_test_group)) > 1:
        cm = confusion_matrix(y_test_group, y_pred_group)
        tpr_group = cm[1, 1] / (cm[1, 0] + cm[1, 1])
        fpr_group = cm[0, 1] / (cm[0, 0] + cm[0, 1])
        precision_group = precision_score(y_test_group, y_pred_group, zero_division=1)

        print(f'Desempenho para o grupo étnico {group}:')
        print(f'TPR: {tpr_group:.2f}')
        print(f'FPR: {fpr_group:.2f}')
        print(f'Precision: {precision_group:.2f}')
        print('-' * 30)
    else:
        print(f'Grupo {group} não possui exemplos positivos e negativos suficientes para calcular a curva ROC.')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('Taxa de Falso Positivo')
plt.ylabel('Taxa de Verdadeiro Positivo')
plt.title('Curva ROC por Grupo Étnico')
plt.legend(title='Grupo Étnico')
plt.grid(True)
plt.show()
```

Alternativas e Complementos ao SMOTE:
ADASYN (Adaptive Synthetic Sampling):

Similar ao SMOTE, mas cria mais exemplos sintéticos para exemplos minoritários que são mais difíceis de aprender, ajustando assim a distribuição do conjunto de dados.
Balanced Random Forest:

Uma variação do Random Forest que ajusta automaticamente os pesos das classes para lidar com desbalanceamento.
Ensemble Methods:

Métodos como EasyEnsemble ou BalanceCascade podem ser utilizados para gerar múltiplos subconjuntos balanceados de dados e treinar modelos sobre estes subconjuntos.
Undersampling:

Reduz o número de exemplos da classe majoritária para balancear o conjunto de dados, embora isso possa levar à perda de informações importantes.
Class Weight Adjustment:

Ajustar os pesos das classes no algoritmo de aprendizado (como foi feito no seu exemplo com RandomForest) para dar mais importância às classes minoritárias.
Técnicas de Mitigação de Viés:
Pre-processing Techniques:

Métodos como Reweighing ou Disparate Impact Remover podem ser usados para ajustar o conjunto de dados antes do treinamento para reduzir o viés.
In-processing Techniques:

Algoritmos como ExponentiatedGradient (usado no seu exemplo) e GridSearchReduction podem ser utilizados para ajustar o modelo durante o treinamento para reduzir o viés.
Post-processing Techniques:

Métodos como EqualizedOddsPostprocessing ou ThresholdOptimizer podem ser aplicados após o treinamento para ajustar as previsões do modelo e reduzir o viés.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.metrics import confusion_matrix, precision_score, roc_curve, roc_auc_score, accuracy_score
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from fairlearn.reductions import ExponentiatedGradient, GridSearch, DemographicParity
from fairlearn.metrics import MetricFrame, selection_rate, false_positive_rate, true_positive_rate
from fairlearn.postprocessing import ThresholdOptimizer
from sklearn.calibration import CalibratedClassifierCV

# Supondo que df_processed já esteja definido
categorical_features = ['sex', 'c_charge_degree', 'r_charge_degree', 'vr_charge_degree', 'c_charge_desc', 'r_charge_desc', 'vr_charge_desc']

# Definir as transformações para colunas categóricas
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combinar as transformações
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features)]
)

df_processed = df_drop_cat_dt.copy()
columns_drop = ['is_recid', 'is_violent_recid']
df_processed = df_processed.drop(columns_drop, axis=1)

# Dividir os dados em X e y
X = df_processed.drop(['two_year_recid'], axis=1)
y = df_processed['two_year_recid']

# Aplicar pré-processamento
X_transformed = preprocessor.fit_transform(X.drop(columns=['race'])).toarray()  # Convertendo para array denso

# Dividir os dados em conjunto de treinamento e teste com amostragem estratificada
X_train, X_test, y_train, y_test, train_index, test_index = train_test_split(
    X_transformed, y, df_processed.index, test_size=0.3, random_state=42, stratify=df_processed['race'])

# Encode sensitive features
sensitive_features_train = df_processed.loc[train_index]['race'].reset_index(drop=True)
label_encoder = LabelEncoder()
sensitive_features_encoded = label_encoder.fit_transform(sensitive_features_train)

# Balancear as classes minoritárias usando SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
sensitive_features_resampled = smote.fit_resample(pd.DataFrame(sensitive_features_encoded), y_train)[0].values.ravel()

# Treinar o Balanced Random Forest Classifier
brf = BalancedRandomForestClassifier(random_state=42)
brf.fit(X_train_resampled, y_train_resampled)

# Calibrar o classificador usando validação cruzada
calibrated_brf = CalibratedClassifierCV(base_estimator=brf, method='sigmoid', cv=5)
calibrated_brf.fit(X_train_resampled, y_train_resampled)

# Avaliar o modelo com métricas de equidade
sensitive_features_test = df_processed.loc[test_index]['race']
y_pred = calibrated_brf.predict(X_test)
metric_frame = MetricFrame(
    metrics={'accuracy': accuracy_score, 'precision': precision_score, 'selection_rate': selection_rate,
             'false_positive_rate': false_positive_rate, 'true_positive_rate': true_positive_rate},
    y_true=y_test,
    y_pred=y_pred,
    sensitive_features=sensitive_features_test
)

print(metric_frame.by_group)

# Avaliar o desempenho para cada grupo étnico
ethnic_groups = df_processed.loc[test_index]['race'].unique()
for group in ethnic_groups:
    group_indices = (df_processed.loc[test_index]['race'] == group)
    X_test_group = X_test[group_indices]
    y_test_group = y_test[group_indices]

    # Fazer previsões para o grupo
    y_pred_prob_group = calibrated_brf.predict_proba(X_test_group)[:, 1]

    if len(set(y_test_group)) > 1:
        fpr, tpr, thresholds = roc_curve(y_test_group, y_pred_prob_group)
        roc_auc = roc_auc_score(y_test_group, y_pred_prob_group)

        plt.plot(fpr, tpr, label=f'{group} (AUC = {roc_auc:.2f})')

        threshold = 0.5
        y_pred_group = (y_pred_prob_group >= threshold).astype(int)
        cm = confusion_matrix(y_test_group, y_pred_group)
        tpr_group = cm[1, 1] / (cm[1, 0] + cm[1, 1])
        fpr_group = cm[0, 1] / (cm[0, 0] + cm[0, 1])
        precision_group = precision_score(y_test_group, y_pred_group)

        print(f'Desempenho para o grupo étnico {group}:')
        print(f'TPR: {tpr_group:.2f}')
        print(f'FPR: {fpr_group:.2f}')
        print(f'Precision: {precision_group:.2f}')
        print('-' * 30)
    else:
        print(f'Grupo {group} não possui exemplos positivos e negativos suficientes para calcular a curva ROC.')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('Taxa de Falso Positivo')
plt.ylabel('Taxa de Verdadeiro Positivo')
plt.title('Curva ROC por Grupo Étnico')
plt.legend(title='Grupo Étnico')
plt.grid(True)
plt.show()

# Usando ExponentiatedGradient para mitigar o viés
expgrad = ExponentiatedGradient(
    estimator=calibrated_brf,
    constraints=DemographicParity(),
    eps=0.01,
    max_iter=500  # Aumentar o número de iterações para explorar mais a fundo
)
expgrad.fit(X_train_resampled, y_train_resampled, sensitive_features=sensitive_features_resampled)
y_pred_expgrad = expgrad.predict(X_test)

# Avaliar o desempenho para ExponentiatedGradient
metric_frame_expgrad = MetricFrame(
    metrics={'accuracy': accuracy_score, 'precision': precision_score, 'selection_rate': selection_rate,
             'false_positive_rate': false_positive_rate, 'true_positive_rate': true_positive_rate},
    y_true=y_test,
    y_pred=y_pred_expgrad,
    sensitive_features=sensitive_features_test
)
print("Desempenho com ExponentiatedGradient:")
print(metric_frame_expgrad.by_group)

# Usando GridSearch para mitigar o viés
grid_search = GridSearch(
    estimator=calibrated_brf,
    constraints=DemographicParity(),
    grid_size=64  # Aumentar o tamanho da grade para melhor exploração
)
grid_search.fit(X_train_resampled, y_train_resampled, sensitive_features=sensitive_features_resampled)
y_pred_grid_search = grid_search.predict(X_test)

# Avaliar o desempenho para GridSearch
metric_frame_grid_search = MetricFrame(
    metrics={'accuracy': accuracy_score, 'precision': precision_score, 'selection_rate': selection_rate,
             'false_positive_rate': false_positive_rate, 'true_positive_rate': true_positive_rate},
    y_true=y_test,
    y_pred=y_pred_grid_search,
    sensitive_features=sensitive_features_test
)
print("Desempenho com GridSearch:")
print(metric_frame_grid_search.by_group)

# Usando ThresholdOptimizer para mitigar o viés
threshold_optimizer = ThresholdOptimizer(
    estimator=calibrated_brf,
    constraints="demographic_parity",
    predict_method='predict_proba',
    prefit=True
)
threshold_optimizer.fit(X_train_resampled, y_train_resampled, sensitive_features=sensitive_features_resampled)
y_pred_threshold_optimizer = threshold_optimizer.predict(X_test, sensitive_features=sensitive_features_test)

# Avaliar o desempenho para ThresholdOptimizer
metric_frame_threshold_optimizer = MetricFrame(
    metrics={'accuracy': accuracy_score, 'precision': precision_score, 'selection_rate': selection_rate,
             'false_positive_rate': false_positive_rate, 'true_positive_rate': true_positive_rate},
    y_true=y_test,
    y_pred=y_pred_threshold_optimizer,
    sensitive_features=sensitive_features_test
)
print("Desempenho com ThresholdOptimizer:")
print(metric_frame_threshold_optimizer.by_group)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, precision_score, roc_curve, roc_auc_score, accuracy_score
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from fairlearn.reductions import ExponentiatedGradient, EqualizedOdds
from fairlearn.metrics import MetricFrame, selection_rate, false_positive_rate, true_positive_rate
from sklearn.calibration import CalibratedClassifierCV
from xgboost import XGBClassifier
import numpy as np

# Supondo que df_processed já esteja definido
categorical_features = ['sex', 'c_charge_degree', 'r_charge_degree', 'vr_charge_degree', 'c_charge_desc', 'r_charge_desc', 'vr_charge_desc']

# Definir as transformações para colunas categóricas
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combinar as transformações
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features)]
)

df_processed = df_drop_cat_dt.copy()
columns_drop = ['is_recid', 'is_violent_recid']
df_processed = df_processed.drop(columns_drop, axis=1)

# Dividir os dados em X e y
X = df_processed.drop(['two_year_recid'], axis=1)
y = df_processed['two_year_recid']

# Aplicar pré-processamento
X_transformed = preprocessor.fit_transform(X.drop(columns=['race'])).toarray()  # Convertendo para array denso

# Dividir os dados em conjunto de treinamento e teste com amostragem estratificada
X_train, X_test, y_train, y_test, train_index, test_index = train_test_split(
    X_transformed, y, df_processed.index, test_size=0.3, random_state=42, stratify=df_processed['race'])

# Encode sensitive features
sensitive_features_train = df_processed.loc[train_index]['race'].reset_index(drop=True)
label_encoder = LabelEncoder()
sensitive_features_encoded = label_encoder.fit_transform(sensitive_features_train)

# Balancear as classes minoritárias usando SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
sensitive_features_resampled = smote.fit_resample(pd.DataFrame(sensitive_features_encoded), y_train)[0].values.ravel()

# Função para ajustar hiperparâmetros com GridSearch
def tune_hyperparameters(X_train, y_train):
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0]
    }
    xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
    grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5, scoring='precision', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    return grid_search.best_estimator_

# Ajustar o modelo com os melhores hiperparâmetros
best_xgb = tune_hyperparameters(X_train_resampled, y_train_resampled)

# Calibrar o classificador usando validação cruzada
calibrated_xgb = CalibratedClassifierCV(estimator=best_xgb, method='sigmoid', cv=5)
calibrated_xgb.fit(X_train_resampled, y_train_resampled)

# Avaliar o classificador calibrado antes da mitigação de viés
y_pred_calibrated = calibrated_xgb.predict(X_test)
metric_frame_calibrated = MetricFrame(
    metrics={'accuracy': accuracy_score, 'precision': precision_score, 'selection_rate': selection_rate,
             'false_positive_rate': false_positive_rate, 'true_positive_rate': true_positive_rate},
    y_true=y_test,
    y_pred=y_pred_calibrated,
    sensitive_features=sensitive_features_test
)
print("Desempenho do classificador calibrado:")
print(metric_frame_calibrated.by_group)

# Usando ExponentiatedGradient com EqualizedOdds
expgrad = ExponentiatedGradient(
    estimator=calibrated_xgb,
    constraints=EqualizedOdds(),
    eps=0.01,
    max_iter=500
)
expgrad.fit(X_train_resampled, y_train_resampled, sensitive_features=sensitive_features_resampled)
y_pred_expgrad = expgrad.predict(X_test)

# Avaliar o desempenho para ExponentiatedGradient com EqualizedOdds
metric_frame_expgrad = MetricFrame(
    metrics={'accuracy': accuracy_score, 'precision': precision_score, 'selection_rate': selection_rate,
             'false_positive_rate': false_positive_rate, 'true_positive_rate': true_positive_rate},
    y_true=y_test,
    y_pred=y_pred_expgrad,
    sensitive_features=sensitive_features_test
)
print("Desempenho com ExponentiatedGradient (EqualizedOdds):")
print(metric_frame_expgrad.by_group)
```
