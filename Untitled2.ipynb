```python
import pandas as pd
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names]

def group_type_crime(description):
    if description == 'NA':
        return 'NA'
    description = description.lower()
    if 'battery' in description or 'assault' in description or 'violence' in description or 'murder' in description or 'batt' in description:
        return 'violence'
    elif 'theft' in description or 'burglary' in description or 'robbery' in description:
        return 'robbery'
    elif 'drug' in description or 'possession' in description or 'trafficking' in description or 'poss' in description or 'cocaine' in description or 'heroin' in description or 'deliver' in description or 'traffick' in description:
        return 'drugs'
    elif 'driving' in description or 'traffic' in description or 'license' in description or 'driv' in description or 'vehicle' in description or 'conduct' in description:
        return 'traffic'
    else:
        return 'other'

def remove_columns(df, columns_to_remove):
    return df.drop(columns=columns_to_remove, errors='ignore')

def string_replace(X):
    if isinstance(X, pd.DataFrame):
        for col in X.columns:
            if X[col].dtype == 'object':
                X[col] = X[col].str.replace('[()]', '', regex=True)
    return X

def custom_group_type_crime(arr):
    vec_group_type_crime = np.vectorize(group_type_crime)
    return vec_group_type_crime(arr)

def add_new_columns(df):
    df['dob_year'] = pd.to_datetime(df['dob'], errors='coerce').dt.year
    df['event_date'] = df['c_offense_date'].combine_first(df['c_arrest_date'])
    df['event_date_year'] = pd.to_datetime(df['event_date'], errors='coerce').dt.year
    df['days_in_jail'] = (pd.to_datetime(df['c_jail_out'], errors='coerce') - pd.to_datetime(df['c_jail_in'], errors='coerce')).dt.days
    return df

def impute_new_columns(df):
    df['days_in_jail'] = df['days_in_jail'].apply(lambda x: df['days_in_jail'].mean() if pd.isna(x) or x < 0 else x)
    df['event_date_year'] = df['event_date_year'].fillna(df['event_date_year'].mean())
    return df

# Definindo o alvo
target = ['two_year_recid']

# Definindo as colunas
categorical_features = ['sex', 'race', 'c_charge_degree', 'c_charge_desc', 'r_charge_degree', 'r_charge_desc', 
                        'vr_charge_degree', 'vr_charge_desc']
numerical_features = ['age', 'juv_fel_count', 'juv_misd_count', 'juv_other_count', 'priors_count', 
                      'dob_year', 'days_in_jail', 'event_date_year']
columns_to_drop = ['id', 'name', 'c_case_number', 'compas_screening_date', 'type_of_assessment', 'decile_score', 'score_text', 'v_type_of_assessment', 
                   'v_decile_score', 'v_score_text', 'is_recid', 'r_case_number', 'r_offense_date', 'vr_case_number', 
                   'vr_offense_date', 'is_violent_recid', 'event_date']

# Pré-processamento de colunas numéricas
numeric_transformer = Pipeline([
    ('selector', DataFrameSelector(numerical_features)),
    ('scaler', StandardScaler())
])

# Pré-processamento de colunas categóricas
categorical_transformer = Pipeline([
    ('selector', DataFrameSelector(categorical_features)),
    ('string_replace', FunctionTransformer(func=string_replace)),
    ('crime_grouper', FunctionTransformer(func=custom_group_type_crime)),    
    ('imputer', SimpleImputer(strategy='constant', fill_value='NA')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combinando pré-processamento
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numerical_features),
    ('cat', categorical_transformer, categorical_features)
])

# Definindo o pipeline completo com o classificador
pipeline = Pipeline([
    ('drop_columns_initial', FunctionTransformer(func=remove_columns, kw_args={'columns_to_remove': columns_to_drop})),
    ('add_new_columns', FunctionTransformer(func=add_new_columns)),
    ('impute_new_columns', FunctionTransformer(func=impute_new_columns)),
    ('drop_datetime_columns', FunctionTransformer(func=remove_columns, kw_args={'columns_to_remove': ['dob', 'c_jail_in', 'c_jail_out', 'c_offense_date', 'c_arrest_date', 'event_date']})),
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(max_depth=3, min_samples_leaf=.03, class_weight="balanced", random_state=42, n_jobs=-1))
])

# Separar os dados de treino e teste
X_train, X_test, y_train, y_test = train_test_split(df.drop('two_year_recid', axis=1), df['two_year_recid'], test_size=0.3, random_state=42)

# Verificar se as colunas foram removidas corretamente
print("Colunas antes do fit:")
print(X_train.columns)

# Treinar o modelo
pipeline.fit(X_train, y_train)

# Fazer previsões e calcular a acurácia
y_pred = pipeline.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia: {accuracy}")
```
