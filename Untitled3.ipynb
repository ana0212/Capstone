```python
import re

def agrupar_tipo_crime(descricao):
    if pd.isna(descricao):
        return 'other'
    descricao = descricao.lower()
    if 'battery' in descricao or 'assault' in descricao or 'violence' in descricao or 'murder' in descricao or 'batt' in descricao:
        return 'violence'
    elif 'theft' in descricao or 'burglary' in descricao or 'robbery' in descricao:
        return 'robbery'
    elif 'drug' in descricao or 'possession' in descricao or 'trafficking' in descricao or 'poss' in descricao or 'cocaine' in descricao or 'heroin' in descricao or 'deliver' in descricao or 'traffick' in descricao:
        return 'drugs'
    elif 'driving' in descricao or 'traffic' in descricao or 'license' in descricao or 'driv' in descricao or 'vehicle' in descricao or 'conduct' in descricao:
        return 'traffic'
    else:
        return 'other'
df_drop['c_charge_desc'] = df_drop['c_charge_desc'].apply(agrupar_tipo_crime)
```

# Análise Bivariada

```python
from sklearn.feature_selection import mutual_info_classif, SelectKBest, chi2
from scipy.stats import chi2_contingency


numerical_cols = ['age', 'juv_fel_count', 'juv_misd_count', 'juv_other_count', 'priors_count', 'event_date_year',
                  'event_date_month', 'event_date_weekday', 'c_jail_in_year', 'c_jail_in_month', 'c_jail_in_weekday']

categorical_cols = ['sex', 'race', 'c_charge_degree', 'c_charge_desc']

df_drop['c_charge_desc'] = pd.Categorical(df_drop['c_charge_desc'], categories=['violence', 'robbery', 'drugs', 'traffic', 'other'])
df_drop['sex'] = df_drop['sex'].astype('category')
df_drop['race'] = df_drop['race'].astype('category')
df_drop['c_charge_degree'] = df_drop['c_charge_degree'].astype('category')

X = df_drop.drop(columns=['is_recid'])
y = df['is_recid']

# Divisão estratificada dos dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

# Análise de Correlação (Pearson) para variáveis numéricas
correlation_matrix = pd.DataFrame(X_train, columns=numerical_cols).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Teste Qui-Quadrado para variáveis categóricas
def chi2_test_of_independence(df, col1, col2):
    cont_table = pd.crosstab(df[col1], df[col2])
    chi2, p, dof, ex = chi2_contingency(cont_table)
    return p

chi2_results = pd.DataFrame(index=categorical_cols, columns=categorical_cols)

for col1 in categorical_cols:
    for col2 in categorical_cols:
        if col1 != col2:
            p_value = chi2_test_of_independence(pd.DataFrame(X_train, columns=X.columns), col1, col2)
            chi2_results.loc[col1, col2] = p_value

print("Chi-Squared Test Results (p-values):")
print(chi2_results)

# One-hot encode as colunas categóricas para usar mutual_info_classif
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Informação Mútua
mi_scores = mutual_info_classif(X_train_processed, y_train, discrete_features='auto')
mi_scores_series = pd.Series(mi_scores, index=preprocessor.get_feature_names_out())

# Plotar Informação Mútua
mi_scores_series.sort_values(ascending=False).plot(kind='barh', figsize=(12, 6))
plt.title('Mutual Information Scores')
plt.show()

# Seleção das Melhores Features usando SelectKBest
# Para variáveis numéricas usando correlação (Pearson)
select_k_best_corr = SelectKBest(score_func=lambda X, y: np.abs(np.corrcoef(X.T, y)[-1, :-1]), k=10)
X_k_best_corr = select_k_best_corr.fit_transform(X_train[numerical_cols], y_train)
best_features_corr = X_train[numerical_cols].columns[select_k_best_corr.get_support()]

print("Best Features by Correlation (Pearson):")
print(best_features_corr)

# Para variáveis categóricas usando Qui-Quadrado
select_k_best_chi2 = SelectKBest(score_func=chi2, k=4)  # Ajustando k para o número de features categóricas disponíveis
X_k_best_chi2 = select_k_best_chi2.fit_transform(X_train[categorical_cols].apply(lambda x: x.cat.codes), y_train)
best_features_chi2 = X_train[categorical_cols].columns[select_k_best_chi2.get_support()]

print("Best Features by Chi-Squared:")
print(best_features_chi2)

# Para todas as variáveis usando Informação Mútua
select_k_best_mi = SelectKBest(score_func=mutual_info_classif, k=10)
X_k_best_mi = select_k_best_mi.fit_transform(X_train_processed, y_train)
best_features_mi = preprocessor.get_feature_names_out()[select_k_best_mi.get_support()]

print("Best Features by Mutual Information:")
print(best_features_mi)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

numerical_cols = ['age', 'juv_fel_count', 'juv_misd_count', 'juv_other_count', 'priors_count', 'c_jail_in_year', 
                  'c_jail_in_month']

categorical_cols = ['sex', 'race', 'c_charge_degree', 'c_charge_desc']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(drop='first'), categorical_cols)
    ])

# Modelos
models = {
    'SVM': SVC(class_weight='balanced', probability=True),
    'Random Forest': RandomForestClassifier(class_weight='balanced'),
    'Balanced Random Forest': BalancedRandomForestClassifier(),
    'XGBoost': XGBClassifier(scale_pos_weight=99)
}

# Função para calcular métricas específicas para a classe de recidivismo (1)
def get_metrics(y_true, y_pred):
    precision = precision_score(y_true, y_pred, pos_label=1)
    recall = recall_score(y_true, y_pred, pos_label=1)
    f1 = f1_score(y_true, y_pred, pos_label=1)
    return precision, recall, f1

# Treinamento e avaliação dos modelos
for name, model in models.items():
    pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', model)])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    
    # Convertendo y_pred para um DataFrame
    y_pred_df = pd.DataFrame(y_pred, index=X_test.index, columns=['prediction'])
    
    print(f"Model: {name}")
    precision, recall, f1 = get_metrics(y_test, y_pred)
    print(f"Overall Metrics for Recidivism (1): Precision: {precision:.2f}, Recall: {recall:.2f}, F1-Score: {f1:.2f}")
    
    # Métricas por raça
    races = X_test['race'].unique()
    race_metrics = []
    for race in races:
        race_indices = X_test[X_test['race'] == race].index
        y_test_race = y_test.loc[race_indices]
        y_pred_race = y_pred_df.loc[race_indices, 'prediction']
        
        precision, recall, f1 = get_metrics(y_test_race, y_pred_race)
        race_metrics.append((race, precision, recall, f1))
        print(f"Metrics for race: {race}")
        print(f"Precision: {precision:.2f}, Recall: {recall:.2f}, F1-Score: {f1:.2f}")
    
    # Verificando se a diferença máxima de precisão entre os grupos é menor que 10%
    precisions = [metric[1] for metric in race_metrics]
    max_diff = max(precisions) - min(precisions)
    print(f"Maximum Precision Difference between Groups: {max_diff:.2f}")
```

# BASELINE 2 (hyperparameter tuning SVM E XGBOOST)
